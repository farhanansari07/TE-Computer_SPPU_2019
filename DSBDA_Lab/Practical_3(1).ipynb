```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("income.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
categorical_col = "age"
numerical_col = "income"
```

```python
summary_stats = df.groupby(categorical_col)[numerical_col].agg(
    mean="mean",
    median="median",
    min="min",
    max="max",
    std="std"
)

summary_stats
```

```python
grouped_values = df.groupby(categorical_col)[numerical_col].apply(list)

grouped_values
```

```python
summary_stats = df.groupby(categorical_col)[numerical_col].agg(
    mean="mean",
    median="median",
    min="min",
    max="max",
    std="std"
)

summary_stats
```

```python
print("Summary Statistics:\n", summary_stats)
print("\nGrouped Values:\n", grouped_values)
```
